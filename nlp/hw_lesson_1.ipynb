```python
col_name = 'processed'
```

```python
def delete_user(df):
    tweet_list = []
    for tw in df.tweet:
        re.findall(r'@[\w]*', tw)
        a = re.sub('@user', ' ', tw)
        tweet_list.append(a.lower())
    df[col_name] = pd.Series(tweet_list, index=df.index)
    
    return df

delete_user(combine_df)
```

```python
def replace_apostrophe(df):
    for num, text in enumerate(df[col_name], 0):
        for word in text.split():
            if word in apostrophe_dict.keys():
                x = text.replace(word, apostrophe_dict[word])
                text = x
                df.iloc[num, 3] = text

    return df
replace_apostrophe(combine_df)
```

```python
def replace_for_full(df):
    for num, text in enumerate(df[col_name], 0):
        for word in text.split():
            if word in short_word_dict.keys():
                x = text.replace(word, short_word_dict[word])
                df.iloc[num, 3] = x

    return df
replace_for_full(combine_df)
```

```python
def replace_emoji_to_word(df):
    for num, text in enumerate(df[col_name], 0):
        for word in text.split():
            if word in emoticon_dict.keys():
                x = text.replace(word, emoticon_dict[word])
                text = x
                df.iloc[num, 3] = x

    return df
replace_emoji_to_word(combine_df)
```

```python
def delete_punc_ssymb_nums(df):
    tweet_list = []
    for tw in df.processed:
        a = re.sub(r'[^\w\s]', ' ', tw)
        a = re.sub(r'[^a-zA-Z0-9]', ' ', tw)
        a = re.sub(r'[^a-zA-Z]', ' ', tw)
        tweet_list.append(a)
    df[col_name] = pd.Series(tweet_list, index=df.index)
    
    return df

delete_punc_ssymb_nums(combine_df)
```

```python
def replace_one_letter_word(df):
    for num, text in enumerate(df[col_name], 0):
        new_sentece = ' '.join([w for w in text.split() if len(w)>1])
        df.iloc[num, 3] = new_sentece

    return df
replace_one_letter_word(combine_df)
```

```python
for num, tw in enumerate(combine_df.processed):
    tokens = nltk.tokenize.word_tokenize(tw)
    combine_df.iat[num, 3] = tokens
```

```python
combine_df['tweet_token'] = combine_df.processed
combine_df.drop('processed', 1, inplace=True)
```

```python
from nltk.corpus import stopwords
mystopwords = stopwords.words('english')

def remove_stopwords(tokens, stopwords = mystopwords):
    for num, sentence in enumerate(tokens):
        
        combine_df.iat[num, 3] = [w for w in sentence if not w in stopwords]
    return combine_df
```

```python
combine_df = remove_stopwords(combine_df.tweet_token)
```

```python
combine_df['tweet_token_clear'] = combine_df.tweet_token
combine_df.drop('tweet_token', 1, inplace=True)
```

```python
stemmer = PorterStemmer()

def proter_stemmer(tokens):
    for num, sentence in enumerate(tokens):
        combine_df.iat[num, 3] = [stemmer.stem(w) for w in sentence]
    return combine_df

combine_df = proter_stemmer(combine_df.tweet_token_clear)
```

```python
combine_df['tweet_stemmed'] = combine_df.tweet_token_clear
combine_df.drop('tweet_token_clear', 1, inplace=True)
```

```python
lemmatizer = WordNetLemmatizer()

def lematize(tokens):
    for num, sentence in enumerate(tokens):
        combine_df.iat[num, 3] = [lemmatizer.lemmatize(w, wordnet.VERB) for w in sentence]
    return combine_df

combine_df = lematize(combine_df.tweet_stemmed)
```

```python
combine_df['tweet_lemmatized'] = combine_df.tweet_stemmed
combine_df.drop('tweet_stemmed', 1, inplace=True)
```
